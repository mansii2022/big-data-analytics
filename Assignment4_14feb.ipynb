```python

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('company_sales_data.csv')
df
```

```python
#question1
plt.plot(df['month_number'])
plt.plot(df['total_profit'])

plt.title('Line Plot')
plt.xlabel('month')
plt.ylabel('profit')
plt.show()



```

```python
#question2
plt.title('multiline plot')

x=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
plt.plot(df[x])
plt.ylabel('total unit sold')
plt.xlabel('month')



```

```python
import matplotlib.pyplot as plt
features = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
values = [10000, 20000, 30000, 40000, 50000,60000]
plt.bar(features, values)
plt.title("Bar Chart of Features")
plt.xlabel("Features")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.show()
```
