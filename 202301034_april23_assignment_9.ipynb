```python
import matplotlib.pyplot as plt
import numpy as np
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
def hypothesis(theta, x):
  return sigmoid(np.dot(theta, x))
from sklearn.datasets import load_iris
iris = load_iris()
x = np.array([5.1, 3.5, 1.4, 0.2]).reshape(1, 4)
theta = np.array([0.5, 0.7, 0.9, 0.1]).reshape(4, 1)
predicted_probability = hypothesis(theta, x)
print("Predicted probability:", predicted_probability)
```
