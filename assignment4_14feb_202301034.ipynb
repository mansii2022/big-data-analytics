```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('company_sales_data.csv')
```

```python
# Question No.1

plt.plot(df['month_number'])
plt.plot(df['total_profit'])

plt.title('line plot')

plt.xlabel('month_number')
plt.ylabel('total_profit')

plt.show()
```

```python
# Question No.2 -Read all product sales data and show it using a multiline plot

plt.plot(df['facecream'],label='facecream')
plt.plot(df['facewash'],label='facewash')
plt.plot(df['toothpaste'],label='toothpaste')
plt.plot(df['bathingsoap'],label='bathingsoap')
plt.plot(df['shampoo'],label='shampoo')
plt.plot(df['moisturizer'],label='moisturizer')

plt.xlabel('month_number')
plt.ylabel('total_sales')

plt.title('Multi-Line Chart')

plt.legend()
plt.show()
```

```python
# Question No.3 -
plt.bar(df['month_number'],df['total_units'])

plt.title('Bar Chart - Month/Total units sold')

plt.xlabel('Month Number')
plt.ylabel('Total Units Sold')

plt.show()
```

```python
# Question No.4

x=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']

plt.scatter(df['month_number'], df['facecream'])
plt.scatter(df['month_number'], df['facewash'])
plt.scatter(df['month_number'], df['toothpaste'])
plt.scatter(df['month_number'], df['bathingsoap'])

plt.title('Scatter Plot')

plt.xlabel('Month_number')
plt.ylabel('x')

plt.show()
```
